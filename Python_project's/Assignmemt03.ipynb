## The "Inventory Guard" Warehouse System 

```python
access_code = "HA14RS12H03"

stock = 500

user = input("Enter Access Code: ")

if user == access_code:

    while True:

        print("1. View Stock")
        print("2. Ship Items")
        print("3. Receive Shipment")
        print("4. Logout")

        choice = input("Enter your choice: ")

        if choice == "1":

            print("Current Stock =", stock)

        elif choice == "2":

            remove_stock = int(input("How many items to ship: "))

            if remove_stock <= stock:

                stock = stock - remove_stock

                print("Updated Stock =", stock)

            else:

                print("Error: Stock shortage")

        elif choice == "3":

            add_stock = int(input("How many items arrived: "))

            stock = stock + add_stock

            print("Updated Stock =", stock)

        elif choice == "4":

            print("Logged Out")

            break

        else:

            print("Invalid Choice")

else:

    print("Access Denied")
```
