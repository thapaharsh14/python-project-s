## Customer Review Analyzer & Report Generator 

```python
## Step 1 - File Create
f = open("raw_reviews.txt","w")
c = f.write( """ "@@@ AMAZING product!!! I bought this last week and it is gr8!!!
delivery was FAST FAST FAST!!!
btw packaging was good but box was damaged :(((  
### worst purchase ever!!! Waste of money!!! waste of 
MONEY!!!  
I love this phone... camera is awesome... battery life is 
AWESOME!!!  
idk why people complain, works fine for me  
screen quality is good but SOUND quality is bad!!!  
LOL this is the best thing I bought this year!!!  
battery drains quickly quickly quickly :(  
Overall I would say it is OKAY product okay product OKAY 
PRODUCT )
""")
f.close()
```

```python
f = open("raw_reviews.txt","r")
c = f.read()
print(c)
f.close()
```

```python
## Task 1 — File Handling 

print(f"First 150 chars:\n",{c[:150]})
print(f"\nLast 150 chars:\n",{c[-150:]})
print(f"\nTotal charaters:\n",len(c))
```

```python
## Task 2 — Basic Cleaning (String Functions Only) 
data = c
data.strip()
data = data.replace("@@@", "")
data = data.replace("###", "")
data = data.replace("!!!", ".")
data = data.replace("...", ".")
data = data.replace(":((", ".")
data = data.replace(":((", "")
data = data.replace(":(", ".")
data = data.replace("(", "")
data = data.replace(")", ".")

data = data.lower()
data = " ".join(data.split())   # remove extra spaces
print(data)
```

```python
## Task 3 — Slang Replacement 

data = data.replace("gr8", "great")
data = data.replace("Btw", "by the way")
data = data.replace("Idk", "i do not know")
data = data.replace("Lol", "funny")

print(data)
```

```python
## Task 4 — Sentence Formatting 

sentences = data.split(".")
sentences = [s.strip().capitalize() for s in sentences if s.strip()]
data = ". ".join(sentences)

print(data)
```

```python
## Task 5 — Duplicate Word Removal 

data = data.replace("fast fast fast", "fast")
data = data.replace("okay product okay product","okay product.")
data = data.replace("quickly quickly quickly", "quickly")

print(data)
```

```python
## Task 6 — Review Analytics (String Analysis) 

## 1. Total number of reviews:

total_reviews = len(data.split("."))
print(f"The total reviews are : {total_reviews}")

## 2. Count how many times these words appear: 

great = data.count("great")
worst = data.count("worst")
awesome = data.count("awesome")
bad = data.count("bad")
good = data.count("good")

print(f"the tatol number of great is {great}")
print(f"the tatol number of worst is {worst}")
print(f"the tatol number of awesome is {awesome}")
print(f"the tatol number of bad is {bad}")
print(f"the tatol number of good is {good}")

##3. Total number of words in cleaned file 

total_word = len(data.split())
print(f"the total number are : {total_word}")
```

```python
## Task 7 — Export Clean Report
f = open("cleaned_reviews_report.txt", "w")
c =  f.write("CLEANED REVIEWS\n")
c =   f.write("-----------------\n")
c = f.write(data + "\n\n")

c = f.write("REPORT\n")
c = f.write("-----------------\n")
c = f.write(f"Total Reviews: {total_reviews}\n")
c = f.write(f"Total Words: {total_word}\n")
c = f.write(f"great count: {great}\n")
c = f.write(f"worst count: {worst}\n")
c = f.write(f"awesome count: {awesome}\n")
c = f.write(f"good count: {good}\n")
c = f.write(f"bad count: {bad}\n")
print(c)
f.close()
```
