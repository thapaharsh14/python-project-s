## Sharma Ji's Canteen System

# 1. Setup Data

```python
menu = {
    "Samosa": 15,
    "Chai": 10,
    "Maggi": 30,
    "Sandwich": 25,
    "Coffee": 20
}

gst = (0.05,)   # GST = 5%

print(f"The Menu of Sharma ji's Canteen is :",)
for item in menu:
    print((item),'=>',menu[item])
```

# 2: Verify Order

```python


def verify_order(order_list):
    valid_items = []

    for item in order_list:
        if item in menu:
            valid_items.append(item)

    return valid_items


order = ["Chai", "Samosa", "Burger", "Maggi"]
result = verify_order(order)

print("Original Order is:",order)
print("valid Order is:",result)
```

# 3: Kitchen Slip

```python

def kitchen_slip(order_list):
    unique_items = set(order_list)

    print("\nKitchen Slip")
    print("------------")

    for item in unique_items:
        print(item)
orders_2 = ["Chai", "Samosa", "Samosa", "Maggi","Chai","Samosa","Maggi","Chai"]
kitchen_slip(orders_2)
```

# 4: Generate Bill

```python
def generate_bill(order_list, menu_dict):
    total = 0

    for item in order_list:
        total += menu_dict[item]

    tax = total * gst[0]
    final_amount = total + tax

    return final_amount
    

orders_1 = ["Chai", "Samosa", "Samosa", "Maggi"]

amount = generate_bill(orders_1 , menu)
print(f"The Final Payable Amount is:",amount)
```

# 5. Special Task: Diwali Offer

```python

offer_menu = {
    item: price - 3
    for item, price in menu.items()
}

print( " --- Diwali Offer Menu --- ")
for item in offer_menu:
    print(item , "=>", offer_menu[item])
```
