```python
# Step 1 : Import Libraries
# Step 2 : Load Dataset
# Step 3 : Data Exploration
# Step 4 : Data Analysis
# Step 5 : Data Visualization
```

```python
import pandas as pd
```

```python
sales_data = pd.read_csv("sales_data.csv")
```

```python
sales_data.head()
```

```python
sales_data.info()
```

```python
sales_data.describe()  # numerical columns ka statistical summary dekhata h.
```

```python
sales_data.shape # Total Rows and Columns
```

```python
# Create Revenue Column

sales_data["Revenue"] = sales_data["Quantity"] * sales_data["Price"] 
```

```python
# Total Revenue

total_revenue = sales_data["Revenue"].sum()
print("Total Revenue =", total_revenue)  
```

```python
# Total Quantity Sold

total_quantity = sales_data["Quantity"].sum()
print("Total Quantity Sold =", total_quantity) 
```

```python
product_revenue = sales_data.groupby("Product")["Revenue"].sum()
product_revenue
```

```python
# Highest product :

highest_product = product_revenue.idxmax()
print("Highest Selling Product :", highest_product)  

highest_revenue = product_revenue.max()
print("Highest Revenue :", highest_revenue)

# max() sabse badi value return karta hai.
# idxmax() us value ka index/label (yahan Product Name) return karta hai.
```

```python
# lowest product :

lowest_product = product_revenue.idxmin()
print("Lowest Selling Product :", lowest_product)

lowest_revenue = product_revenue.min()
print("Lowest Revenue :", lowest_revenue)
```

```python
# average revenue :
average_revenue = sales_data["Revenue"].mean()

print("Average Revenue :", average_revenue)
```

```python
# Count of Orders:

total_orders = sales_data["Order_ID"].count()
print("Total Orders :", total_orders)
```

```python
# unique products :
unique_products = sales_data["Product"].unique()
print(unique_products)
```

```python
# Category-wise Order Count :
sales_data["Category"].value_counts()
```

## Operator:

```python
# Show only Laptop orders
sales_data[sales_data["Product"] == "Laptop"]
```

```python
# Show products whose price is greater than 10000
sales_data[sales_data["Price"] > 10000]
```

```python
sales_data[sales_data["Quantity"] < 3]
```

```python
sales_data[
    (sales_data["Product"] == "Laptop") &
    (sales_data["Price"] > 50000)
]
```

```python
# Top 5 Highest Revenue
sales_data.sort_values("Revenue", ascending=False).head(5) 
```

```python
# Bottom 5 Revenue
sales_data.sort_values("Revenue").head(5)
```

## Matplotlib :

```python
import matplotlib.pyplot as plt
```

```python
product_revenue = sales_data.groupby("Product")["Revenue"].sum()
product_revenue.plot(kind="bar")

plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Revenue")

plt.show()
```

```python
top5 = product_revenue.sort_values(ascending=False).head(5)

top5.plot(kind="pie",startangle=90)

plt.title("Top 5 Product Revenue Share")
plt.ylabel("")
plt.show()
```

```python
# Top 5 Product revenues

top5 = product_revenue.sort_values(ascending=False).head()

top5.plot(kind="bar")

plt.title("Top 5 Products")
plt.show()
```

```python
# bottom 5 product revenue
bottom5 = product_revenue.sort_values().head()

bottom5.plot(kind="bar", color="red")

plt.title("Bottom 5 Products")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.scatter(sales_data["Quantity"],sales_data["Revenue"])

plt.title("Quantity vs Revenue")
plt.xlabel("Quantity")
plt.ylabel("Revenue")

plt.show()
```

```python
# ==========================
# Business Insights
# ==========================

print("=" * 40)
print("      BUSINESS INSIGHTS")
print("=" * 40)

# Total Revenue
total_revenue = sales_data["Revenue"].sum()
print("1. Total Revenue :", total_revenue)

# Average Revenue
average_revenue = sales_data["Revenue"].mean()
print("2. Average Revenue :", round(average_revenue, 2))

# Total Quantity Sold
total_quantity = sales_data["Quantity"].sum()
print("3. Total Quantity Sold :", total_quantity)

# Top Selling Product
top_product = sales_data.groupby("Product")["Revenue"].sum().idxmax()
top_revenue = sales_data.groupby("Product")["Revenue"].sum().max()

print("4. Top Selling Product :", top_product)
print("   Revenue :", top_revenue)

# Lowest Selling Product
low_product = sales_data.groupby("Product")["Revenue"].sum().idxmin()
low_revenue = sales_data.groupby("Product")["Revenue"].sum().min()

print("5. Lowest Selling Product :", low_product)
print("   Revenue :", low_revenue)

print("\nConclusion:")
print("- The highest revenue is generated by", top_product)
print("- The lowest revenue is generated by", low_product)
print("- The company should focus more on high-performing products.")
print("- Sales strategies can be improved for low-performing products.")
```
