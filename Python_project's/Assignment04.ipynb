## 2. Dictionaries (Key–Value)

```python
# Create dictionary
prices = {
    'iPhone': 50000,
    'Samsung': 40000,
    'Oppo': 30000,
    'OnePlus': 45000
}

print(prices)
```

```python
# Update iPhone price
prices['iPhone'] = 48000
```

```python
# Add Pixel price
prices['Pixel'] = 42000

print("Updated Prices:", prices)
```

## 3. Sets (Union & Intersection)

```python
morning_visitors = ['Aman', 'Ravi', 'Sita', 'Rahul']
evening_visitors = ['Ravi', 'Sita', 'Ankit', 'Neha']

# Convert to sets
morning_set = set(morning_visitors)
evening_set = set(evening_visitors)

# Union (unique customers)
unique_customers = morning_set | evening_set
print("Unique Customers:", unique_customers)

# Intersection (common customers)
common_customers = morning_set & evening_set
print("Visited Both Time:", common_customers)
```

## 4. Comprehensions (Flash Sale Discount) 

```python
# Dictionary comprehension (10% discount)
discounted_prices = {k: v * 0.9 for k, v in prices.items()}
print("Discounted Prices:", discounted_prices)

# List comprehension (only values)
discount_list = [v * 0.9 for v in prices.values()]
print("Discounted Price List:", discount_list)
```

## 5. Tuples (Immutable Data Structure) 

```python
customer_name = "Ravi"
phone = "iPhone"
final_price = discounted_prices[phone]

# Store as tuple
bill = (customer_name, phone, final_price)

print("Customer Bill:", bill)
```
