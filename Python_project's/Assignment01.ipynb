## What the Assignment whants?
- Take salary input for 5 employees.
- Convert thoso inputs to numbers.
- Calculate allowances, gross pay and netpay using arithmetic.
- Evaluat salary categories using only comparison + logical Opertors

```python
## 1 Input Capture:

name = input("Enter Employee Name: ")
base = float(input("Enter Base Pay: "))


## from xyz Company-provided percentages:

hra_percent = 20
da_percent = 15 
pf_percent = 12 

## 2 Data Transformation:

hra = (hra_percent / 100) * base
da = (da_percent / 100) * base
pf = (pf_percent / 100) * base


## 3 The Arithmetic Engine:

gross = base + hra + da
net_pay = gross - pf

## 4 Logical Evaluation:

is_high_earner = net_pay > 80000
is_mid_range = (net_pay >= 40000) and (net_pay <= 80000)
is_low = net_pay < 40000

print("\nEmployee:", name)
print("Net Pay:", net_pay)

print("High Earner:", is_high_earner)
print("Mid Range:", is_mid_range)
print("Low Earner:", is_low)
```
